```python
from umpyre.data.scraper import BaseballReferenceScraper
import pandas as pd

# initialize scraper
scraper = BaseballReferenceScraper()
```

```python
year = 2015
al_url = "http://www.baseball-reference.com/leagues/AL/%d.shtml" % year
nl_url = "http://www.baseball-reference.com/leagues/NL/%d.shtml" % year

al = scraper.data_from_url(al_url)
nl = scraper.data_from_url(nl_url)
```

```python
standings = pd.DataFrame([])

def combine_standings(l):
    standings = pd.concat([l['standings_W'],l['standings_C'],l['standings_E']])
    standings.replace('--',0,inplace=True)
    return standings

for year in range(1999,2016):
    print "Processing year %d" % year
# year = 2015
    al_url = "http://www.baseball-reference.com/leagues/AL/%d.shtml" % year
    nl_url = "http://www.baseball-reference.com/leagues/NL/%d.shtml" % year

    al = scraper.data_from_url(al_url)
    nl = scraper.data_from_url(nl_url)

    al_standings = combine_standings(al)
    nl_standings = combine_standings(nl)
    nl_standings['year'] = year
    nl_standings['league'] = 'NL'
    al_standings['year'] = year
    al_standings['league'] = 'NL'

    standings = pd.concat([standings, al_standings,nl_standings])
    
standings
```
